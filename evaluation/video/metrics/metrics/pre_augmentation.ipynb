```python
import os
import pandas as pd

qwen_csv_path = "/mnt/SSD3/lina/ssb/vlm_inference/Qwen2.5-VL-32B-Instruct/Task1_Qwen2.5-VL-32B-Instruct_all.csv"
internvl_csv_path = "/mnt/SSD3/lina/ssb/vlm_inference/InternVL3_5-38B/Task1_InternVL3_5-38B_all.csv"
name_map_csv_path = "/home/lina/icassp2026/ICASSP2026/evaluation/audio/new_old_name_map.csv"
```

```python
qwen_csv = pd.read_csv(qwen_csv_path)
internvl_csv = pd.read_csv(internvl_csv_path)
# name_map_csv = pd.read_csv(name_map_csv_path)

qwen_csv = qwen_csv.loc[:, ~qwen_csv.columns.str.contains('justification')]
internvl_csv = internvl_csv.loc[:, ~internvl_csv.columns.str.contains('justification')]


display(qwen_csv.head())
display(internvl_csv.head())
# display(name_map_csv.head())
```

```python
# Import the merge_segments function
from merge_segments import merge_segments
import tempfile
import os

# Create temporary directory for all files
temp_dir = tempfile.mkdtemp()

# Process InternVL data
print("=" * 60)
print("PROCESSING INTERNVL DATA")
print("=" * 60)

# Create temporary files for InternVL data
internvl_temp_path = os.path.join(temp_dir, 'internvl_predictions.csv')

# Save the InternVL DataFrame to temporary CSV file
internvl_csv.to_csv(internvl_temp_path, index=False)

# Merge InternVL segments
internvl_merged = merge_segments(internvl_temp_path, verbose=True)

# Display the first few rows of merged InternVL data
print("\nMerged InternVL data:")
print("=" * 40)
display(internvl_merged.head())

# Process Qwen data
print("\n" + "=" * 60)
print("PROCESSING QWEN DATA")
print("=" * 60)

# Create temporary files for Qwen data
qwen_temp_path = os.path.join(temp_dir, 'qwen_predictions.csv')

# Save the Qwen DataFrame to temporary CSV file
qwen_csv.to_csv(qwen_temp_path, index=False)

# Merge Qwen segments
qwen_merged = merge_segments(qwen_temp_path, verbose=True)

# Display the first few rows of merged Qwen data
print("\nMerged Qwen data:")
print("=" * 40)
display(qwen_merged.head())

# Clean up temporary files
os.remove(internvl_temp_path)
os.remove(qwen_temp_path)
os.rmdir(temp_dir)

print("\n" + "=" * 60)
print("MERGE COMPLETED FOR ALL MODELS")
print("=" * 60)
```

```python
name_mapping_csv_path = '/mnt/SSD1/prateik/icassp_vlm/new_old_name_map.csv'
# Read the name mapping CSV
name_map_df = pd.read_csv(name_mapping_csv_path)


# Create a dictionary for quick mapping
name_map = dict(zip(name_map_df.iloc[:, 0], name_map_df.iloc[:, 1]))

# Function to apply the mapping
def map_filename(filename):
    # Remove _segment_X.mp4 from the filename before mapping
    base_name = filename.split('_segment_')[0]
    # Get the mapped name, if not found return original
    mapped_name = name_map.get(base_name, base_name)
    return mapped_name

# Apply the mapping to both dataframes
print("Applying name mapping to InternVL data...")
internvl_merged['file_name'] = internvl_merged['file_name'].apply(map_filename)

print("\nApplying name mapping to Qwen data...")
qwen_merged['file_name'] = qwen_merged['file_name'].apply(map_filename)

# Display the first few rows of each dataframe to verify the mapping
print("\nInternVL data with mapped names:")
print("=" * 40)
display(internvl_merged.head())

print("\nQwen data with mapped names:")
print("=" * 40)
display(qwen_merged.head())
```

```python
# Import the calculate_feature_metrics function
from calculate_metrics import calculate_feature_metrics
import tempfile
import os

# Create temporary directory for all files
temp_dir = tempfile.mkdtemp()

# ============================================================================
# CALCULATE METRICS FOR INTERNVL DATA
# ============================================================================
print("=" * 80)
print("CALCULATING METRICS FOR INTERNVL DATA")
print("=" * 80)

# Create temporary file for InternVL data
internvl_temp_path = os.path.join(temp_dir, 'internvl_predictions.csv')

# Save the merged and mapped InternVL DataFrame to temporary CSV file
internvl_merged.to_csv(internvl_temp_path, index=False)

print("Calculating metrics for InternVL data...")
print("="*50)

# Calculate metrics for InternVL data
internvl_metrics = calculate_feature_metrics(
    predictions_csv=internvl_temp_path,
    experiment_name="internvl_analysis",
    verbose=True
)

# Display InternVL results
print("\nInternVL Metrics Results:")
print("="*40)
display(internvl_metrics)

# ============================================================================
# CALCULATE METRICS FOR QWEN DATA
# ============================================================================
print("\n" + "=" * 80)
print("CALCULATING METRICS FOR QWEN DATA")
print("=" * 80)

# Create temporary file for Qwen data
qwen_temp_path = os.path.join(temp_dir, 'qwen_predictions.csv')

# Save the merged and mapped Qwen DataFrame to temporary CSV file
qwen_merged.to_csv(qwen_temp_path, index=False)

print("Calculating metrics for Qwen data...")
print("="*50)

# Calculate metrics for Qwen data
qwen_metrics = calculate_feature_metrics(
    predictions_csv=qwen_temp_path,
    experiment_name="qwen_analysis",
    verbose=True
)

# Display Qwen results
print("\nQwen Metrics Results:")
print("="*40)
display(qwen_metrics)

# Clean up temporary files
os.remove(internvl_temp_path)
os.remove(qwen_temp_path)
os.rmdir(temp_dir)

print("\n" + "=" * 80)
print("METRICS CALCULATION COMPLETED FOR ALL MODELS")
print("=" * 80)
```
