```python
import os
import pandas as pd

csv_dir = 'output'
csv_files = [f for f in os.listdir(csv_dir) if f.endswith('.csv')]

crop = [x for x in csv_files if 'crop' in x.lower()]
pos = [x for x in csv_files if 'pos' in x.lower()]
base = [x for x in csv_files if 'crop' not in x.lower() and 'pos' not in x.lower()]

print("BASE")
for file in base:
    print(file)
    df = pd.read_csv(os.path.join(csv_dir, file))
    display(df.head())
print("CROP")
for file in crop:
    df = pd.read_csv(os.path.join(csv_dir, file))
    display(df.head())
print("POSE")
for file in pos:
    df = pd.read_csv(os.path.join(csv_dir, file))
    display(df.head())
```

```python
# Read the base models data
base_dfs = []
for file in base:
    df = pd.read_csv(os.path.join(csv_dir, file))
    base_dfs.append(df)

# Read crop and pose data
crop_dfs = []
for file in crop:
    df = pd.read_csv(os.path.join(csv_dir, file))
    crop_dfs.append(df)

pose_dfs = []
for file in pos:
    df = pd.read_csv(os.path.join(csv_dir, file))
    pose_dfs.append(df)

# Get the first dataframe from each list (since we only have one file of each type)
base_df = base_dfs[0]
crop_df = crop_dfs[0]
pose_df = pose_dfs[0]

# Create comparison dataframes for crop and pose features
crop_features = ['blank_stare', 'close_eyes', 'eye_blinking', 'oral_automatisms', 
                 'face_pulling', 'face_twitching', 'head_turning']
pose_features = ['arm_flexion', 'occur_during_sleep', 'tonic', 'clonic', 
                'arm_straightening', 'figure4', 'limb_automatisms']

metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']

# Function to create comparison table
def create_comparison_table(base_df, comparison_df, features, model_name):
    comparison_data = []
    
    for metric in metrics:
        row = {'Metric': metric}
        for feature in features:
            base_val = base_df[base_df['Metric'] == metric][feature].values[0]
            comp_val = comparison_df[comparison_df['Metric'] == metric][feature].values[0]
            row[f'{feature} (Base)'] = base_val
            row[f'{feature} ({model_name})'] = comp_val
            row[f'{feature} (Diff)'] = comp_val - base_val
        comparison_data.append(row)
    
    return pd.DataFrame(comparison_data)

# Create and display crop comparison
print("Crop Features Comparison (Base vs Crop)")
crop_comparison = create_comparison_table(base_df, crop_df, crop_features, 'Crop')
display(crop_comparison)

print("\nPose Features Comparison (Base vs Pose)")
pose_comparison = create_comparison_table(base_df, pose_df, pose_features, 'Pose')
display(pose_comparison)

# Add a summary of improvements
def summarize_improvements(comparison_df, features, model_name):
    print(f"\nSummary of {model_name} Improvements:")
    for feature in features:
        diff_col = f'{feature} (Diff)'
        f1_diff = comparison_df[comparison_df['Metric'] == 'F1 Score'][diff_col].values[0]
        acc_diff = comparison_df[comparison_df['Metric'] == 'Accuracy'][diff_col].values[0]
        
        print(f"\n{feature}:")
        print(f"  F1 Score change: {f1_diff:.3f}")
        print(f"  Accuracy change: {acc_diff:.3f}")

print("\nPerformance Summary:")
summarize_improvements(crop_comparison, crop_features, 'Crop')
summarize_improvements(pose_comparison, pose_features, 'Pose')
```
